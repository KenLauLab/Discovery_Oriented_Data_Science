```python
# imports the pcreode package
import pcreode
# matplotlib is a commonly used package for plotting
import matplotlib.pyplot as plt
# pandas is a package used for making the handling of large data sets easier 
import pandas as pd
# numpy is very common package for handling arrays and matrices
import numpy as np
# this line tells matplotlib to plot results inline with the current jupyter notebook
%matplotlib inline
```

### Create a nonstructured dataset

```python
num_pts = 5000
uniform_x = np.random.uniform(0, 100, size=num_pts)
uniform_y = np.random.uniform(0, 100, size=num_pts)
uniform = np.vstack( ( uniform_x, uniform_y)).T
```

```python
plt.scatter( uniform[:,0], uniform[:,1])
```

### Create a simple structured data set with similar dimensions of the nonstructured data set

```python
mid_npts = 500
mid_x = np.random.uniform(  0, 50, mid_npts)
mid_y = np.random.normal( 50, 5, mid_npts)    #uniform( 40, 60, mid_npts)
mid   = np.vstack( ( mid_x, mid_y)).T
```

```python
top_npts = 500
top_x = np.random.uniform( 50, 100, top_npts)
top_y = top_x + np.random.normal( 0, 5, top_npts) #uniform( -10, 10, top_npts)
top   = np.vstack( ( top_x, top_y)).T
```

```python
bot_npts = 500
bot_x = np.random.uniform( 50, 100, bot_npts)
bot_y = -bot_x + 100 + np.random.normal( 0, 5, bot_npts) #uniform( -10, 10, bot_npts)
bot   = np.vstack( ( bot_x, bot_y)).T
```

```python
noise_npts = 50
noise_x = np.random.uniform( 25, 75, noise_npts)
noise_y = np.random.uniform( 25, 75, noise_npts)
noise   = np.vstack( ( noise_x, noise_y)).T
```

```python
struct = np.vstack( ( mid, top, bot, noise))
```

```python
dens = pcreode.Density( struct)
best = dens.radius_best_guess()
density = dens.get_density( best, mute=True)
```

```python
plt.scatter( struct[:,0], struct[:,1], c=density)
```

```python
data = pd.DataFrame( struct, columns=['X','Y'])
data.to_csv( "../data/synthetic.csv", index=False)
```
